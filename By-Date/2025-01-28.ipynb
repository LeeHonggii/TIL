# 1과목 - 소프트웨어공학

### 소프트웨어
- 컴퓨터 시스템을 구성하는 프로그램, 절차, 규칙, 관련 문서들의 총칭입니다. 하드웨어를 실행하고 제어하기 위한 모든 프로그램을 포함합니다.

소프트웨어 특징 - 상품성, 복잡성, 변경가능성, 복제성

### 시스템 
- 컴퓨터로 처리가능한 자료를 입력하고 처리하는걸 말함

### 소프트웨어 재공학 
- 소프트웨어 재공학은 기존의 노후화된 시스템을 새로운 시스템으로 재구축하는 과정입니다.


##### 주요 목적
* 소프트웨어 품질을 전반적으로 향상시키고 시스템의 안정성을 확보하기 위함입니다
* 시스템의 유지보수를 더욱 용이하게 만들어 향후 관리 비용을 절감할 수 있습니다
* 전체적인 시스템 성능을 개선하여 더 효율적인 운영이 가능해집니다
* 새로운 비즈니스 요구사항을 수용할 수 있는 유연한 구조로 변경합니다

과정 : 분석- 구성 -역공학 - 이식

### CASE(Computer Aided Software Engineering)

##### 정의
CASE는 소프트웨어 개발 전 과정에 걸쳐 자동화된 도구들을 활용하여 개발 생산성과 품질을 향상시키는 방법론입니다.

##### CASE의 주요 특징
* 소프트웨어 개발의 생산성을 향상시킵니다
* 체계적이고 표준화된 개발 방법을 제공합니다
* 신뢰성 있는 소프트웨어를 개발할 수 있도록 지원합니다
* 유지보수가 용이한 소프트웨어 개발을 가능하게 합니다

##### CASE 도구의 분류

##### 1. 상위 CASE(Upper CASE)
* 계획 수립, 요구분석, 설계 단계를 지원
* 데이터 흐름도(DFD), ER 다이어그램 작성 도구
* 자동 문서 생성기
* 프로토타입 작성기

##### 2. 하위 CASE(Lower CASE)
* 구현, 테스트, 유지보수 단계를 지원
* 프로그램 생성기
* 디버거
* 테스트 도구
* 형상 관리 도구

##### 3. 통합 CASE(I-CASE)
* 상위 CASE와 하위 CASE를 통합
* 소프트웨어 개발 전 단계를 일관성 있게 지원
* 저장소(Repository)를 통한 정보 공유

##### CASE 도구의 장점
* 소프트웨어 품질 향상
* 개발 기간 단축
* 유지보수 비용 절감
* 프로젝트 표준화 용이
* 작업 자동화를 통한 생산성 향상

##### CASE 도구의 단점
* 도구 구매 및 교육에 많은 비용 소요
* 새로운 도구 사용에 대한 거부감
* 기존 개발 방법의 변경 필요
* 도구 간의 호환성 문제 발생 가능

##### CASE 도구 선정 시 고려사항
* 조직의 개발 환경과의 적합성
* 사용 용이성과 학습의 편의성
* 하드웨어 및 소프트웨어 요구사항
* 벤더의 지원 능력과 신뢰성
* 투자 대비 효과 분석

### 소프트웨어 생명주기 
- 소프트웨어 생명주기는 소프트웨어를 개발하기 위해 정의하고 운용하는 과정을 체계적으로 관리하는 방법론입니다

## 소프트웨어 생명주기 모형의 특징
* 개발 단계별 산출물이 명확하게 정의됩니다
* 각 단계의 결과가 검증됩니다
* 체계적인 접근 방법을 제공합니다
* 프로젝트 진행 상황을 관리자가 파악하기 쉽습니다

## 주요 생명주기 모형

### 1. 폭포수 모형(Waterfall Model)
* 가장 오래되고 폭넓게 사용된 전통적인 모형
* 각 단계가 순차적으로 진행
* 단계별 산출물이 명확함
* 이전 단계로 돌아가기 어려움

#### 개발 단계
1. 요구사항 분석
2. 설계
3. 구현
4. 테스트
5. 유지보수

### 2. 프로토타입 모형(Prototype Model)
* 사용자의 요구사항을 정확히 파악하기 위한 모형
* 시제품을 만들어 사용자 의견 반영
* 빠른 피드백 가능
* 완성될 시스템의 기능을 미리 확인 가능

### 3. 나선형 모형(Spiral Model)
* 위험 분석을 강조한 모형
* 점진적으로 개발하는 반복적 접근방법
* 프로토타이핑을 포함한 여러 소프트웨어 개발방법론의 장점을 혼합

#### 4단계 반복
1. 계획 수립
2. 위험 분석
3. 개발 및 검증
4. 고객 평가

### 4. 반복적 모형(Iterative Model)
* 개발 과정의 반복을 통해 완성도를 높임
* 위험 요소의 조기 발견 가능
* 고객 요구사항의 변경 수용이 용이
* 테스트가 초기부터 수행됨

## 소프트웨어 생명주기 모형 선택 시 고려사항
* 프로젝트의 규모와 복잡도
* 개발 조직의 환경
* 고객의 요구사항 명확성
* 프로젝트의 위험도
* 개발 일정과 비용
* 개발팀의 기술 수준

## 각 모형의 적용 상황
* 폭포수 모형: 요구사항이 명확하고 변경이 적은 프로젝트
* 프로토타입 모형: 사용자 인터페이스가 중요한 프로젝트
* 나선형 모형: 대규모이며 위험도가 높은 프로젝트
* 반복적 모형: 요구사항 변경이 잦은 프로젝트

### 하향식 설계와 상향식 설계

### 하향식 설계(Top-Down Design)

### 정의
* 전체 시스템을 구성하는 상위 모듈에서 시작하여 하위 모듈로 세분화하는 설계 방식
* 큰 문제를 작은 문제로 분할하면서 설계를 구체화함

### 특징
* 추상화 수준이 높은 상태에서 시작
* 단계적 상세화(Stepwise Refinement) 방식 사용
* 구조적 프로그래밍에 적합
* 통합 테스트가 용이함

### 장점
* 전체적인 시스템 구조를 빠르게 파악 가능
* 프로젝트 초기에 전체 구조 설계 가능
* 각 기능의 모듈화가 용이
* 중복되는 코드를 최소화할 수 있음

### 단점
* 하위 수준의 세부사항이 늦게 구체화됨
* 초기 설계 시 많은 시간이 소요
* 시스템 자원에 대한 고려가 늦어질 수 있음

### 상향식 설계(Bottom-Up Design)

### 정의
* 가장 하위 수준의 모듈부터 설계를 시작하여 상위 모듈을 구성하는 방식
* 기본 모듈을 먼저 구현하고 이를 결합하여 상위 모듈을 완성

### 특징
* 실제 구현 가능한 모듈부터 시작
* 재사용 가능한 모듈을 먼저 개발
* 객체지향 프로그래밍에 적합
* 단위 테스트가 용이함

### 장점
* 하위 모듈의 재사용성이 높음
* 실제 구현이 빠르게 진행됨
* 기능 검증이 초기부터 가능
* 시스템 자원의 효율적 활용이 용이

### 단점
* 전체 시스템 구조를 늦게 파악할 수 있음
* 상위 수준의 모듈 통합 시 어려움 발생 가능
* 전체 시스템 최적화가 어려울 수 있음

### 설계 방식 선택 기준

### 하향식 설계 적용 상황
* 새로운 시스템을 처음부터 개발할 때
* 시스템의 전체 구조가 명확할 때
* 구조적 프로그래밍을 적용할 때
* 큰 규모의 프로젝트에서 업무 분담이 필요할 때

### 상향식 설계 적용 상황
* 재사용 가능한 컴포넌트 개발이 중요할 때
* 시스템의 세부 기능이 명확할 때
* 객체지향 프로그래밍을 적용할 때
* 기존 모듈의 재사용이 많은 프로젝트

## 혼합 설계 방식
* 실제 프로젝트에서는 두 가지 방식을 혼합하여 사용하는 것이 일반적
* 프로젝트의 특성과 요구사항에 따라 적절한 비율로 조합
* 각 방식의 장점을 최대한 활용하고 단점을 보완

# HIPO(Hierarchy plus Input-Process-Output)

## 정의
HIPO는 시스템의 기능을 표현하기 위한 문서화 도구이며, 하향식 소프트웨어 개발을 위한 문서화 기법입니다.

## HIPO의 특징
* 기능과 자료의 계층 관계를 시각적으로 표현
* 하향식 개발 방법을 지원
* 기능 중심의 문서화 지원
* 시스템의 전체적인 기능을 파악하기 쉬움

## HIPO 구성도의 종류

### 1. 가시적 도표(Visual Table of Contents)
* 시스템의 기능을 계층적으로 표현
* 전체 시스템의 기능과 흐름을 한눈에 파악 가능
* 상위 기능과 하위 기능의 관계를 트리 구조로 표현
* 기능의 분할과 계층 구조를 보여줌

### 2. 총체적 도표(Overview Diagram)
* 프로그램의 기능을 입력, 처리, 출력 형태로 표현
* 기능에 대한 전반적인 정보 제공
* 입력된 데이터의 처리 과정을 기술
* 처리된 결과가 어떤 형태로 출력되는지 표현

### 3. 세부적 도표(Detail Diagram)
* 총체적 도표에 표현된 기능의 세부적인 처리 내용 표현
* 입력, 처리, 출력에 대한 상세한 정보 제공
* 실제 프로그램의 처리 과정을 상세하게 표현
* 각 기능의 알고리즘을 구체적으로 기술

## HIPO의 장점
* 문서화가 용이하고 이해하기 쉬움
* 유지보수가 용이함
* 시스템의 구조적 분석이 가능
* 시스템 설계의 분할과 정복 접근이 가능
* 변경 및 수정이 용이

## HIPO의 단점
* 시스템의 전체적인 이해가 필요
* 중복된 정보를 표현할 수 있음
* 시간에 따른 변화를 표현하기 어려움
* 비용과 시간이 많이 소요될 수 있음


# V-모델(V-Model)

## 정의
V-모델은 소프트웨어 개발 프로세스를 검증 단계와 함께 표현한 생명주기 모델로, 폭포수 모델을 확장한 형태입니다.

## 특징
* 각 개발 단계와 테스트 단계가 1:1로 매핑됨
* 개발 단계와 테스트 단계가 V자 형태로 대칭을 이룸
* 검증과 확인에 중점을 둔 모델
* 개발 과정과 테스트 과정의 상호 연관성을 강조

## V-모델의 구성 단계

### 왼쪽 부분 (개발 단계)
1. 요구사항 분석
   * 사용자의 요구사항을 수집하고 분석
   * 시스템의 기능과 제약사항 정의

2. 시스템 설계
   * 전체 시스템 구조 설계
   * 시스템 컴포넌트 정의

3. 아키텍처 설계
   * 시스템의 주요 컴포넌트 설계
   * 컴포넌트 간의 인터페이스 정의

4. 상세 설계
   * 각 모듈의 상세 설계
   * 알고리즘과 데이터 구조 정의

5. 구현(코딩)
   * 실제 프로그램 코드 작성
   * 단위 모듈 구현

### 오른쪽 부분 (테스트 단계)
1. 운영 테스트/인수 테스트
   * 사용자 요구사항 충족 여부 검증
   * 시스템의 실제 운영 환경 검증

2. 시스템 테스트
   * 전체 시스템의 기능 검증
   * 비기능적 요구사항 검증

3. 통합 테스트
   * 컴포넌트 간 인터페이스 검증
   * 컴포넌트 통합 기능 검증

4. 단위 테스트
   * 개별 모듈의 기능 검증
   * 모듈 단위의 오류 검출

## V-모델의 장점
* 체계적인 테스트 단계 제공
* 개발과 테스트의 관계가 명확
* 각 단계별 산출물이 명확
* 프로젝트 관리와 품질 관리가 용이

## V-모델의 단점
* 개발 중간에 요구사항 변경이 어려움
* 각 단계가 완료된 후에야 다음 단계로 진행 가능
* 프로토타입 개발이 어려움
* 초기에 많은 비용과 시간이 소요

## V-모델 적용 시 고려사항
* 요구사항이 명확하고 변경이 적은 프로젝트에 적합
* 충분한 테스트 자원과 시간이 필요
* 철저한 문서화가 요구됨
* 개발팀의 높은 전문성이 요구됨

## 테스트 전략 수립
* 각 개발 단계별 테스트 계획 수립
* 테스트 케이스와 시나리오 개발
* 테스트 환경과 도구 준비
* 테스트 수행 절차와 기준 정의

# 에자일 개발 방법론(Agile Methodology)

## 정의
에자일 방법론은 변화하는 요구사항에 유연하게 대응하면서 반복적이고 점진적으로 소프트웨어를 개발하는 방법론입니다.

## 에자일 선언문의 핵심 가치
* 프로세스와 도구보다 **개인과 상호작용** 중시
* 포괄적인 문서보다 **작동하는 소프트웨어** 중시
* 계약 협상보다 **고객과의 협력** 중시
* 계획을 따르기보다 **변화에 대응하기** 중시

## 주요 특징
* 반복적이고 점진적인 개발
* 작은 단위의 릴리스
* 지속적인 피드백 반영
* 팀원 간의 긴밀한 협력
* 고객과의 지속적인 소통
* 유연한 요구사항 수용

## 대표적인 에자일 방법론

### 1. 스크럼(Scrum)
* 스프린트(Sprint) 단위로 개발 진행
* 제품 백로그(Product Backlog) 관리
* 데일리 스크럼 미팅 진행
* 스프린트 리뷰와 회고 실시

#### 주요 역할
* 제품 소유자(Product Owner)
* 스크럼 마스터(Scrum Master)
* 개발팀(Development Team)

### 2. XP(eXtreme Programming)
* 페어 프로그래밍
* 지속적인 통합
* 테스트 주도 개발(TDD)
* 리팩토링
* 간단한 디자인

### 3. 칸반(Kanban)
* 작업 시각화
* 진행 중인 작업 제한
* 작업 흐름 관리
* 지속적인 개선

## 에자일의 장점
* 빠른 피드백과 대응
* 고객 만족도 향상
* 팀 생산성 향상
* 위험 요소 조기 발견
* 품질 향상

## 에자일의 단점
* 예측이 어려움
* 문서화가 부족할 수 있음
* 팀원의 높은 숙련도 요구
* 대규모 프로젝트에 적용이 어려움

## 에자일 도입 시 고려사항
* 조직 문화의 변화 필요
* 팀원들의 교육과 훈련
* 적절한 도구와 환경 구축
* 고객의 적극적인 참여 필요

## 에자일 프랙티스

* 점진적인 도입과 개선

* 팀의 권한 부여
* 잦은 제품 출시
* 현재 비즈니스 요구에 집중
* 반복적이고 점진적인 개발
* 모든 변경은 가역적

* 설계와 구축 반복(Design and Build Iteration)

### 5가지 프로세스
1. 전반적인 모델 개발

3. 기능별 계획 수립
4. 기능별 설계
5. 기능별 구축

### 주요 특징

* 객체 모델링 강조

# DBMS(Database Management System)

## 정의
DBMS는 사용자와 데이터베이스 사이에서 사용자의 요구에 따라 정보를 생성해주고 데이터베이스를 관리해주는 소프트웨어입니다.

## DBMS의 필수 기능

### 1. 정의 기능(Definition)
* 데이터베이스 구조 정의
* 데이터 형식, 구조, 제약조건 명세
* 저장 방법 정의
* 데이터의 물리적/논리적 모델 정의

### 2. 조작 기능(Manipulation)
* 데이터 검색, 갱신, 삽입, 삭제
* 사용자와 데이터베이스 사이의 인터페이스 제공
* 데이터베이스 접근 및 조작 방법 제공
* 데이터의 복구 및 보안 기능

### 3. 제어 기능(Control)
* 데이터 무결성 유지
* 데이터 접근 제어
* 병행 제어(Concurrency Control)
* 보안 및 권한 관리

## DBMS의 장단점

### 장점
* 데이터 중복 최소화
* 데이터 일관성 유지
* 데이터 무결성 보장
* 데이터 보안 강화
* 데이터 독립성 확보
* 표준화 용이
* 데이터 관리 효율성 증가

### 단점
* 비용이 많이 듦
* 백업과 회복이 어려움
* 전문가 필요
* 과부하 시 복잡함
* 시스템이 복잡함

## DBMS의 종류

### 1. 계층형 DBMS
* 트리 구조 사용
* 상하 종속적 관계
* 1:N 관계 표현
* IBM의 IMS가 대표적

### 2. 네트워크형 DBMS
* 그래프 구조 사용
* 종속적 관계 및 순환 관계 표현
* N:M 관계 표현
* IDS가 대표적

### 3. 관계형 DBMS
* 테이블 기반 구조
* SQL 사용
* Oracle, MySQL, MS-SQL 등
* 현재 가장 많이 사용됨

### 4. 객체지향형 DBMS
* 객체 개념 도입
* 멀티미디어 데이터 처리
* 복잡한 데이터 표현 가능
* 객체 관계형 DBMS로 발전

## DBMS 선정 기준
* 기술 지원
* 성능 및 기능
* 가용성 및 편의성
* 상호 호환성
* 보안성
* 확장성
* 비용

## 주요 DBMS 제품
* Oracle
* MySQL
* MS-SQL
* PostgreSQL
* MongoDB
* MariaDB
* IBM DB2

## DBMS 관련 주요 용어
* 테이블(Table)
* 필드(Field)
* 레코드(Record)
* 키(Key)
* 스키마(Schema)
* 트랜잭션(Transaction)
* 인덱스(Index)
* 뷰(View)

# 요구사항 명세 기법(Requirements Specification Techniques)

## 정의
요구사항 명세는 요구사항을 분석한 후 이를 문서화하는 기법으로, 시스템 개발의 기준이 되는 문서를 작성하는 기법입니다.

## 주요 명세 기법

### 1. 정형 명세 기법
* 수학적 기호와 정형화된 표기법 사용
* 모호성이 없는 정확한 명세 가능
* 요구사항의 자동 분석 가능
* 고신뢰성이 요구되는 시스템에 적합

#### 대표적인 정형 명세 언어
* Z
* VDM
* Petri-Net
* SDL

### 2. 비정형 명세 기법
* 자연어를 기반으로 한 명세
* 이해하기 쉽고 작성이 용이
* 다양한 다이어그램 활용
* 일반적인 시스템 개발에 적합

#### 주요 표현 방법
* 자연어 명세
* 구조적 명세
* 객체지향 명세

### 3. 반정형 명세 기법
* 정형 명세와 비정형 명세의 혼합

# UML(Unified Modeling Language)

## 정의
UML은 시스템 개발 과정에서 산출물을 명세화, 시각화, 문서화할 때 사용되는 표준화된 모델링 언어입니다.

## UML 다이어그램 종류

### 1. 구조적 다이어그램(Structural Diagrams)

#### 클래스 다이어그램(Class Diagram)
* 클래스와 클래스 간의 관계 표현
* 속성과 메서드 정의
* 상속, 연관, 집합, 의존 관계 표현
* 시스템의 정적 구조 표현

#### 객체 다이어그램(Object Diagram)
* 클래스의 인스턴스 표현
* 객체 간의 관계 표현
* 시스템의 특정 시점 상태 표현

#### 컴포넌트 다이어그램(Component Diagram)
* 물리적 컴포넌트 표현
* 컴포넌트 간의 의존 관계 표현
* 시스템 구성요소 표현

#### 배치 다이어그램(Deployment Diagram)
* 시스템의 물리적 아키텍처 표현
* 하드웨어, 소프트웨어 노드 표현
* 컴포넌트의 물리적 위치 표현

#### 패키지 다이어그램(Package Diagram)
* 시스템의 패키지 구조 표현
* 패키지 간의 의존 관계 표현

### 2. 행위적 다이어그램(Behavioral Diagrams)

#### 유스케이스 다이어그램(Use Case Diagram)
* 사용자와 시스템 간의 상호작용 표현
* 시스템의 기능적 요구사항 표현
* 액터와 유스케이스의 관계 표현

#### 시퀀스 다이어그램(Sequence Diagram)
* 객체 간의 상호작용을 시간 순서대로 표현
* 메시지 흐름 표현
* 동적 행위 표현

#### 상태 다이어그램(State Diagram)
* 객체의 상태 변화 표현
* 이벤트에 따른 상태 전이 표현
* 객체의 생명주기 표현

#### 활동 다이어그램(Activity Diagram)
* 업무 처리 과정 표현
* 알고리즘의 논리적 절차 표현
* 병행 처리 표현

## UML 관계 표현

### 1. 연관 관계(Association)
* 클래스 간의 참조 관계
* 양방향/단방향 관계
* 다중성(Multiplicity) 표현

### 2. 집합 관계(Aggregation)
* 전체와 부분의 관계
* 느슨한 결합 관계
* 공유 가능한 관계

### 3. 합성 관계(Composition)
* 강한 집합 관계
* 생명주기 종속 관계
* 배타적 소유 관계

### 4. 일반화 관계(Generalization)
* 상속 관계
* 특수화/일반화 관계
* IS-A 관계

### 5. 의존 관계(Dependency)
* 일시적인 사용 관계
* 약한 결합 관계
* 변경 영향 관계

## UML 확장 메커니즘

### 1. 스테레오타입(Stereotype)
* 기존 UML 요소의 의미 확장
* <<stereotype>> 형태로 표현

### 2. 태그값(Tagged Value)
* 속성 추가 정의
* {tag=value} 형태로 표현

### 3. 제약조건(Constraint)
* 규칙이나 조건 명시
* {constraint} 형태로 표현

# 럼바우 객체지향 분석 기법(Rumbaugh's Object-Oriented Analysis)

## 정의
럼바우(Rumbaugh) 방법론은 객체 모델링 기법(OMT: Object Modeling Technique)이라고도 하며, 시스템의 정적 구조와 동적 행위를 모델링하는 객체지향 분석 방법론입니다.

## 주요 모델

### 1. 객체 모델(Object Model)
* 시스템의 정적 구조 표현
* 객체와 객체 간의 관계 정의
* 클래스와 속성, 연산 정의

#### 주요 표기법
* 클래스 다이어그램
* 객체 다이어그램
* 관계(연관, 집합, 일반화) 표현
* 다중성(Multiplicity) 표현

### 2. 동적 모델(Dynamic Model)
* 시스템의 시간적 행위와 제어 흐름 표현
* 객체들의 상태 변화와 상호작용 표현
* 이벤트와 상태 변화 정의

#### 주요 표기법
* 상태 다이어그램
* 이벤트 흐름도
* 상태 전이표
* 시나리오

### 3. 기능 모델(Functional Model)
* 시스템의 데이터 처리 과정 표현
* 입력값이 출력값으로 변환되는 과정 표현
* 데이터 흐름과 연산 정의

#### 주요 표기법
* 자료 흐름도(DFD)
* 제약 조건
* 연산 명세
* 최적화 표현

## 분석 절차

### 1. 문제 정의
* 요구사항 수집
* 문제 범위 설정
* 목표 설정
* 제약사항 파악

### 2. 객체 모델 구축
* 클래스와 객체 식별
* 관계 정의
* 속성 정의
* 연산 정의

### 3. 동적 모델 구축
* 사건(이벤트) 식별
* 상태 식별
* 상태 전이 정의
* 글로벌 상태도 작성

### 4. 기능 모델 구축
* 입출력 값 식별
* 기능(함수) 식별
* 제약조건 식별
* 데이터 흐름도 작성

### 5. 모델 검증
* 3개 모델 간의 일관성 검증
* 요구사항 충족 여부 검증
* 완전성 검증
* 중복성 검증

## 장점
* 체계적인 분석 방법 제공
* 세 가지 관점의 완전한 모델링 가능
* 이해하기 쉬운 표기법 사용
* 점진적인 개발 가능

## 단점
* 세 가지 모델 작성에 많은 시간 소요
* 모델 간 일관성 유지가 어려움
* 소규모 프로젝트에는 과도한 작업량
* 실시간 시스템 분석에 한계

## 적용 시 고려사항
* 프로젝트 규모와 특성 고려
* 팀의 객체지향 이해도
* 도구 지원 여부
* 일정과 비용 제약
