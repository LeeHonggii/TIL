# UML 다이어그램의 구조적/행위적 분류

## 구조적 다이어그램(Structural Diagrams)
시스템을 구성하는 요소들의 정적인 구조와 관계를 표현

### 1. 클래스 다이어그램(Class Diagram)
* 시스템의 클래스들과 그들 간의 관계 표현
* 속성(Attributes)과 메서드(Methods) 포함
* 상속, 연관, 집합, 의존 관계 등을 표현
* 시스템의 정적인 구조를 가장 잘 표현하는 다이어그램

### 2. 객체 다이어그램(Object Diagram)
* 클래스의 인스턴스(객체) 표현
* 특정 시점의 객체 상태 표현
* 클래스 다이어그램의 특정 시점 스냅샷

### 3. 컴포넌트 다이어그램(Component Diagram)
* 시스템의 컴포넌트 구조 표현
* 컴포넌트 간의 의존성 표현
* 물리적 소프트웨어 구성요소 표현

### 4. 배치 다이어그램(Deployment Diagram)
* 시스템의 물리적 구조 표현
* 하드웨어, 소프트웨어 노드 배치 관계
* 실행 환경의 구성 표현

### 5. 패키지 다이어그램(Package Diagram)
* 시스템의 패키지 구조 표현
* 패키지 간의 의존 관계 표현
* 큰 시스템의 구조적 조직화

## 행위적 다이어그램(Behavioral Diagrams)
시스템의 동적인 행위와 상호작용을 표현

### 1. 유스케이스 다이어그램(Use Case Diagram)
* 시스템과 사용자 간의 상호작용 표현
* 시스템의 기능적 요구사항 표현
* 액터와 유스케이스 관계 표현
* 시스템의 범위 정의

### 2. 시퀀스 다이어그램(Sequence Diagram)
* 객체 간의 시간 순서에 따른 상호작용 표현
* 메시지 교환 순서 표현
* 객체의 생명선과 활성화 표현

### 3. 상태 다이어그램(State Diagram)
* 객체의 상태 변화 표현
* 이벤트에 따른 상태 전이 표현
* 객체의 생명주기 표현

### 4. 활동 다이어그램(Activity Diagram)
* 업무 처리 과정이나 절차 표현
* 병행 처리와 조건 분기 표현
* 알고리즘의 논리적 흐름 표현

### 5. 통신 다이어그램(Communication Diagram)
* 객체 간의 관계와 메시지 교환 표현
* 객체의 구조적 관계에 중점
* 시퀀스 다이어그램의 대안적 표현

### 6. 타이밍 다이어그램(Timing Diagram)
* 객체의 상태 변화를 시간에 따라 표현
* 실시간 시스템의 타이밍 제약 표현
* 시간 기반의 제약사항 모델링

## 다이어그램 선택 기준
* 표현하고자 하는 시스템 측면
* 프로젝트의 단계
* 이해관계자의 요구
* 시스템의 복잡도
* 개발 방법론

# 유스케이스 다이어그램(Use Case Diagram)

## 정의
유스케이스 다이어그램은 시스템과 사용자(액터) 간의 상호작용을 표현하는 다이어그램으로, 시스템의 기능적 요구사항을 나타냅니다.

## 기본 구성요소

### 1. 액터(Actor)
* 시스템과 상호작용하는 외부 개체
* 사람, 외부 시스템, 시간 등
* 스틱맨(막대 사람) 모양으로 표현
* 액터 이름은 명사형으로 작성

### 2. 유스케이스(Use Case)
* 시스템이 제공하는 기능
* 타원형으로 표현
* 동사형으로 작성
* 사용자 관점에서 의미있는 단위 작업

### 3. 시스템 경계(System Boundary)
* 사각형으로 표현
* 시스템의 범위 정의
* 유스케이스들을 포함
* 시스템 이름 표시

## 관계(Relationships)

### 1. 연관 관계(Association)
* 액터와 유스케이스 간의 상호작용
* 실선으로 표현
* 양방향 또는 단방향 가능
* 가장 기본적인 관계

### 2. 포함 관계(Include)
* 반드시 포함되어야 하는 관계
* <<include>> 로 표시
* 점선 화살표 사용
* 공통적인 기능을 분리할 때 사용

### 3. 확장 관계(Extend)
* 조건에 따라 확장되는 관계
* <<extend>> 로 표시
* 점선 화살표 사용
* 선택적인 기능을 분리할 때 사용

### 4. 일반화 관계(Generalization)
* 상속 관계를 표현
* 실선 화살표(삼각형) 사용
* 액터 간 또는 유스케이스 간에 사용
* 공통된 특성을 추상화할 때 사용

## 작성 지침

### 1. 액터 식별
* 시스템 사용자 파악
* 외부 시스템 파악
* 시간이나 이벤트 파악
* 주액터와 부액터 구분

### 2. 유스케이스 식별
* 액터별 목표 파악
* 시스템 기능 도출
* 비즈니스 가치 고려
* 적절한 추상화 수준 유지

### 3. 관계 설정
* 적절한 관계 유형 선택
* 불필요한 관계 제거
* 복잡도 관리
* 가독성 고려

## 작성 시 고려사항
* 사용자 관점에서 작성
* 적절한 추상화 수준 유지
* 간결하고 명확하게 표현
* 비즈니스 가치 중심
* 명명 규칙 준수

## 유스케이스 명세서 작성
* 유스케이스 이름
* 액터
* 선행조건
* 후행조건
* 기본 흐름
* 대체 흐름
* 예외 흐름
* 비기능적 요구사항

# 소프트웨어 아키텍처(Software Architecture)

## 정의
소프트웨어 아키텍처는 시스템의 기본 구조와 구성요소, 그들 간의 관계를 정의하는 설계 체계입니다.

## 주요 아키텍처 패턴

### 1. 계층화 패턴(Layered Pattern)
* 기능을 계층으로 구분
* 상위 계층은 하위 계층에 의존
* 관심사의 분리 구현
* 대표적 예: OSI 7계층

#### 주요 계층 구성
* 프레젠테이션 계층
* 비즈니스 계층
* 퍼시스턴스 계층
* 데이터베이스 계층

### 2. 클라이언트-서버 패턴(Client-Server Pattern)
* 서비스 요청자와 제공자 구조
* 서버의 자원 공유 가능
* 분산 시스템 구현에 적합
* 네트워크 기반 서비스에 활용

### 3. MVC 패턴(Model-View-Controller)
* 모델: 데이터와 비즈니스 로직
* 뷰: 사용자 인터페이스
* 컨트롤러: 사용자 입력 처리
* 관심사 분리를 통한 유지보수성 향상

### 4. 파이프-필터 패턴(Pipe-Filter Pattern)
* 데이터 스트림 처리
* 연속적인 처리 단계
* 필터의 재사용성
* 데이터 변환 작업에 적합

### 5. 마이크로서비스 아키텍처
* 작은 독립적 서비스들의 조합
* 서비스별 독립적 배포 가능
* 높은 확장성과 유연성
* DevOps 문화와 잘 어울림

## 아키텍처 품질 속성

### 1. 가용성(Availability)
* 시스템 운영 가능 시간
* 장애 복구 능력
* 중복성과 내결함성

### 2. 성능(Performance)
* 응답 시간
* 처리량
* 자원 사용률

### 3. 보안(Security)
* 기밀성
* 무결성
* 가용성
* 인증과 인가

### 4. 확장성(Scalability)
* 수평적 확장
* 수직적 확장
* 부하 분산

### 5. 유지보수성(Maintainability)
* 코드 가독성
* 모듈화
* 테스트 용이성

## 아키텍처 설계 원칙

### 1. 관심사의 분리(Separation of Concerns)
* 기능별 모듈화
* 책임의 명확한 구분
* 복잡도 관리

### 2. 단일 책임 원칙(Single Responsibility)
* 하나의 컴포넌트는 하나의 책임만
* 변경 사유의 단일화
* 응집도 향상

### 3. 개방-폐쇄 원칙(Open-Closed)
* 확장에는 열려있음
* 수정에는 닫혀있음
* 유연한 설계

## 아키텍처 문서화

### 1. 뷰 모델
* 논리적 뷰
* 프로세스 뷰
* 물리적 뷰
* 개발 뷰

### 2. 문서화 항목
* 아키텍처 결정사항
* 컴포넌트 구조
* 인터페이스 정의
* 배포 구성
* 품질 속성 시나리오
